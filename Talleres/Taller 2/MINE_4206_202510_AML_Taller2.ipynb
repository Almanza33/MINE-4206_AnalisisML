### **Problemática**:
La acuicultura, en particular la producción de tilapia, ha experimentado un crecimiento significativo en algunas regiones de Colombia, convirtiéndose en una fuente importante de proteína animal y sustento económico para muchas comunidades. Sin embargo, este sector enfrenta varios desafíos relacionados con la calidad del agua y el bienestar de los peces, lo que puede tener un impacto directo en la sostenibilidad del ecosistema acuático y la rentabilidad de las operaciones.

En muchos sistemas acuícolas, las fluctuaciones en parámetros críticos como el oxígeno disuelto, el pH, la temperatura y la turbidez del agua pueden desencadenar condiciones adversas para los peces, afectando su salud, crecimiento y tasa de supervivencia. Estos factores, cuando no se gestionan adecuadamente, pueden llevar a pérdidas económicas sustanciales y a un impacto ambiental negativo debido al uso ineficiente de recursos y a la contaminación del agua.

La implementación de tecnologías de monitoreo basadas en Internet de las Cosas (IoT) ha permitido la recolección continua de datos sobre la calidad del agua y la salud de los peces en tiempo real. Sin embargo, a pesar de la abundancia de datos, los acuicultores a menudo carecen de herramientas adecuadas para analizar y predecir cómo estos parámetros interactúan y afectan la productividad y la sostenibilidad ambiental de sus operaciones.
El objetivo de este proyecto es desarrollar soluciones, basadas en machine learning, que, utilizando datos de monitoreo IoT, permita predecir y optimizar la calidad del agua y la salud de la tilapia en estanques de acuicultura. Este modelo no solo podría ayudar a los acuicultores a anticipar problemas relacionados con la calidad del agua y la salud de los peces, sino que también puede facilitar una gestión más eficiente de los recursos acuáticos, promoviendo la sostenibilidad ambiental. Al integrar la predicción de variables clave el proyecto busca mejorar la eficiencia en el uso del agua, reducir el riesgo de contaminación y contribuir a la conservación de los ecosistemas acuáticos.
